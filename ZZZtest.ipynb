```python
def identifier_cles_manquantes(system_data_keys: list, max_prim: int = 2280, max_sec: int = 3):
    """
    Identifie les cls manquantes dans system_data.
    
    Args:
        system_data (dict): Le dictionnaire contenant les donnes
        max_prim (int): ID principal maximum (inclusif). Dfaut 2280.
        max_sec (int): ID secondaire maximum (inclusif). Dfaut 3.
        
    Returns:
        list: La liste des cls manquantes tries.
    """
    # 1. Gnrer l'ensemble thorique complet (toutes les combinaisons possibles)
    # range(max + 1) car range est exclusif sur la fin
    toutes_cles_possibles = {f"{p}-{s}" for p in range(max_prim + 1) for s in range(max_sec + 1)}
    
    # 2. Rcuprer l'ensemble des cls existantes
    cles_existantes = set(system_data_keys)
    
    # 3. Faire la diffrence (ce qui est prvu MOINS ce qui existe)
    cles_manquantes = toutes_cles_possibles - cles_existantes
    
    # 4. Fonction de tri pour avoir "2-1" avant "10-1" (tri numrique)
    def tri_numerique(cle):
        prim, sec = map(int, cle.split('-'))
        return (prim, sec)
    
    resultat_trie = sorted(list(cles_manquantes), key=tri_numerique)
    
    # 5. Affichage pour le notebook
    print(f"--- RAPPORT DE DIAGNOSTIC ---")
    print(f"Total thorique : {len(toutes_cles_possibles)} combinaisons")
    print(f"Total prsent   : {len(cles_existantes)} cls")
    print(f"Total manquant  : {len(resultat_trie)} cls")
    print("-" * 30)
    
    if len(resultat_trie) == 0:
        print(" Aucune cl manquante. Le dataset est complet.")
    else:
        print(" Exemples de cls manquantes (10 premires) :")
        print(resultat_trie[:10])
        if len(resultat_trie) > 10:
            print("...")
            
    return resultat_trie

# --- UTILISATION ---
# Remplacez 'system_data' par votre variable relle
# manquants = identifier_cles_manquantes(system_data)
print(identifier_cles_manquantes(system_data_keys=system_data_keys))
```
